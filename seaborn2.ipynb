```python
import numpy as num
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

# FacetGrid

```python
sns.FacetGrid(tips)
```

```python
sns.FacetGrid(tips,row='smoker',col='time')
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.hist,'total_bill')
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.hist,'total_bill',color='green')
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.hist,'total_bill',color='orange',bins=100)
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(sns.regplot,'total_bill','tip')
```

```python
x=sns.FacetGrid(tips,hue='smoker',col='time')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x=sns.FacetGrid(tips,hue='smoker',col='time')
x=x.map(plt.scatter,'total_bill','tip')
x=x.add_legend()
```

```python
x=sns.FacetGrid(tips,col='day',row='smoker')
x=x.map(sns.boxplot,'total_bill','time')
```

```python
x=sns.FacetGrid(tips,col='day',row='smoker')
x=x.map(sns.barplot,'total_bill','time',palette='summer')
```

```python
sns.relplot(x='tip',y='total_bill',data=tips)
plt.show()
```

```python
sns.regplot(x='tip',y='total_bill',data=tips,color='red')
```
