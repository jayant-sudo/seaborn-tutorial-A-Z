# Pair Grid


```python
#This code will only create the structure ...
iris=sns.load_dataset("iris")
iris=sns.load_dataset("iris")
x=sns.PairGrid(iris)
#This will visualise data into the graph
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)  # it will put scatter on non-diagonal....
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue="species")
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue="species",palette="winter_r")
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue="species",palette="autumn")
x=x.map_diag(plt.hist,histtype='step',linewidth=2)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,vars=["petal_length","petal_width"])
x=x.map_diag(plt.hist)
```

```python
x=sns.PairGrid(iris,vars=["petal_length","petal_width"],hue="species")
x=x.map_diag(plt.hist,edgecolor='gray')
x=x.map_offdiag(plt.scatter,edgecolor="black")
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,x_vars=["petal_length","petal_width"],y_vars=["sepal_length","sepal_width"])
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_upper(plt.scatter)
x=x.map_lower(sns.kdeplot)
```

```python
x=sns.PairGrid(iris,hue="species")
x=x.map_diag(plt.hist)
x=x.map_upper(plt.scatter)
x=x.map_lower(sns.kdeplot)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue="species",hue_kws={'marker':['D','s','+']})
x=x.map(plt.scatter,s=30)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue="species",hue_kws={'marker':['D','s','+']},palette="winter_r")
x=x.map(plt.scatter,s=30,edgecolor="black")
x=x.add_legend()
```
